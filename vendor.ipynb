```python
def analyze_vendors_by_item(item_name, hospital_name=None):
    """
    Analyzes vendor data to find the lowest item rate vendor and top 10 vendors for a specific item.

    Args:
        item_name (str): The name of the item to analyze.
        hospital_name (str, optional): The name of the hospital to filter by.
                                       Defaults to None (analyze across all hospitals).

    Returns:
        tuple: A tuple containing:
            - pandas.DataFrame: The vendor(s) with the lowest 'Item Rate'.
            - pandas.DataFrame: The top 10 vendors based on 'Item Rate'.
    """
    # Filter by item name
    filtered_df_item = df2[df2['Item'] == item_name].copy()

    # Further filter by hospital name if provided
    if hospital_name:
        filtered_df_final = filtered_df_item[filtered_df_item['Hospital'] == hospital_name].copy()
    else:
        filtered_df_final = filtered_df_item.copy()

    # Find the minimum 'Item Rate'
    min_item_rate = filtered_df_final['Item Rate'].min()

    # Identify the vendor(s) with the lowest 'Item Rate'
    lowest_rate_vendor = filtered_df_final[filtered_df_final['Item Rate'] == min_item_rate]

    # Sort by 'Item Rate' and select top 10 unique vendors
    top_10_vendors = filtered_df_final.sort_values(by='Item Rate', ascending=True).drop_duplicates(subset=['Name']).head(10)

    return lowest_rate_vendor, top_10_vendors
```

```python
# Get a list of all unique item names
unique_items = df2['Item'].unique()

# Create an empty dictionary to store the results for each item
all_items_analysis_results = {}

# Loop through each unique item name and analyze vendors
for item in unique_items:
    lowest_rate, top_10 = analyze_vendors_by_item(item)
    all_items_analysis_results[item] = {
        'lowest_rate_vendor': lowest_rate,
        'top_10_vendors': top_10
    }

# Display a sample of the stored results for the first item
if unique_items.size > 0:
    first_item = unique_items[0]
    print(f"Analysis results for item: {first_item}")
    display(all_items_analysis_results[first_item]['lowest_rate_vendor'])
    display(all_items_analysis_results[first_item]['top_10_vendors'])
else:
    print("No unique items found in the data.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Redefine the on_button_click function with enhanced display
def on_button_click(b):
    clear_output(wait=True) # Clear previous output

    # Re-display the dropdowns and button at the top
    display(hospital_dropdown)
    display(item_dropdown)
    display(parent_category_dropdown)
    display(vendor_dropdown)
    display(analyze_button)

    selected_hospital = hospital_dropdown.value
    selected_item = item_dropdown.value
    selected_parent_category = parent_category_dropdown.value
    selected_vendor = vendor_dropdown.value

    # Start with the full df2 and apply filters using boolean indexing
    filtered_df_items = df2.copy()

    # Apply filters based on selections
    if selected_hospital != 'All Hospitals':
        filtered_df_items = filtered_df_items[filtered_df_items['Hospital'] == selected_hospital]

    if selected_parent_category != 'All Categories':
        filtered_df_items = filtered_df_items[filtered_df_items['Item Parent Category'] == selected_parent_category]

    if selected_vendor != 'All Vendors':
        filtered_df_items = filtered_df_items[filtered_df_items['Name'] == selected_vendor]

    # Update item dropdown options based on current hospital, parent category, and vendor selections
    available_items = sorted(filtered_df_items['Item'].unique().tolist())
    available_items = ['All Items'] + available_items if available_items else ['No Items Available']

    current_selected_item = item_dropdown.value
    item_dropdown.options = available_items

    if current_selected_item in item_dropdown.options:
        item_dropdown.value = current_selected_item
    elif item_dropdown.options:
        item_dropdown.value = item_dropdown.options[0]
    else:
        item_dropdown.value = 'No Items Available'

    # Apply the final item filter only if 'All Items' is not selected and a valid item is selected
    if item_dropdown.value and item_dropdown.value != 'No Items Available' and item_dropdown.value != 'All Items':
         filtered_df_final = filtered_df_items[filtered_df_items['Item'] == item_dropdown.value]
    else:
        filtered_df_final = filtered_df_items # No need for another copy here

    display(widgets.HTML(f"<h3>Analysis for Item: {item_dropdown.value}, Hospital: {selected_hospital}, Parent Category: {selected_parent_category}, and Vendor: {selected_vendor}</h3>"))

    if filtered_df_final.empty:
        display(widgets.HTML("<b>No data available for the selected criteria.</b>"))
    else:
        if selected_vendor != 'All Vendors':
            display(widgets.HTML(f"<b>Items provided by {selected_vendor}:</b>"))
            unique_items_by_vendor = filtered_df_final.drop_duplicates(subset=['Item'])
            item_list_html = "<ul>"
            for index, row in unique_items_by_vendor.iterrows():
                item_list_html += f"<li>Item: {row['Item']}, Total Value: {row['Total Value']:.2f}</li>"
            item_list_html += "</ul>"
            display(widgets.HTML(item_list_html))
        else:
            min_item_rate = filtered_df_final['Item Rate'].min()
            lowest_rate_vendor = filtered_df_final[filtered_df_final['Item Rate'] == min_item_rate]

            top_10_vendors = filtered_df_final.sort_values(by='Item Rate', ascending=True).drop_duplicates(subset=['Name']).head(10)

            # Enhanced styling for lowest rate vendor
            def highlight_lowest_rate(s):
                is_lowest = s == s.min()
                return ['background-color: lightgreen' if v else '' for v in is_lowest]

            display(widgets.HTML("<b>--- Vendor with the Lowest Item Rate ---</b>"))
            if not lowest_rate_vendor.empty:
                # Apply styling to highlight the lowest Item Rate specifically
                display(lowest_rate_vendor[['Name', 'Hospital', 'Item Parent Category', 'Item', 'Total Value', 'Item Rate']].reset_index(drop=True).style.apply(highlight_lowest_rate, subset=['Item Rate']).set_properties(**{'font-size': '12pt', 'white-space': 'pre-wrap'}))
            else:
                display(widgets.HTML("<i>No vendor found with the lowest item rate for this selection.</i>"))

            display(widgets.HTML("<b>--- Top 10 Vendors by Item Rate (Lowest to Highest) ---</b>"))
            if not top_10_vendors.empty:
                 # Display top 10 vendors as a styled table
                 display(top_10_vendors[['Name', 'Hospital', 'Item Parent Category', 'Item', 'Total Value', 'Item Rate']].reset_index(drop=True).style.background_gradient(subset=['Item Rate'], cmap='Greens').set_properties(**{'font-size': '12pt', 'white-space': 'pre-wrap'}))

                 # Add a bar chart for top 10 vendors by Item Rate for better comparison
                 plt.figure(figsize=(10, 6))
                 sns.barplot(x='Item Rate', y='Name', data=top_10_vendors.sort_values(by='Item Rate', ascending=False), palette='viridis')
                 plt.title(f'Top 10 Vendors by Item Rate for {item_dropdown.value}')
                 plt.xlabel('Item Rate')
                 plt.ylabel('Vendor Name')
                 plt.tight_layout()
                 plt.show() # Use plt.show() to display the plot

            else:
                display(widgets.HTML("<i>No top 10 vendors found for this selection.</i>"))


# Link the on_button_click function to the button's on_click method
analyze_button.on_click(on_button_click)

# Initially call the on_button_click function to display results for the default selections
on_button_click(None)
```

```python
%%writefile notebook.md
# Operations Dashboard: Vendor and Item Analysis

This dashboard provides insights into vendor performance and item pricing based on various filters. Use the interactive controls to explore data for specific hospitals, item categories, items, and vendors.

## How to Use the Dashboard

1.  **Select Filters:** Use the dropdown menus at the top to select criteria for your analysis:
    *   **Select Hospital:** Choose a specific hospital or 'All Hospitals' to analyze data across all locations.
    *   **Select Parent Category:** Narrow down the analysis to items within a specific parent category or select 'All Categories'.
    *   **Select Vendor:** Focus on a particular vendor or select 'All Vendors' to compare vendors.
    *   **Select Item:** Choose a specific item. Note that the available items in this dropdown will update based on your Hospital, Parent Category, and Vendor selections. Select 'All Items' to see aggregated results for the applied filters.

2.  **Analyze Data:** After making your selections, click the **Analyze Data** button. The dashboard will update to display the relevant analysis based on your chosen filters.

## Understanding the Outputs

The dashboard provides different views depending on whether you have selected a specific vendor or are analyzing across all vendors ('All Vendors').

### When 'All Vendors' is Selected:

*   **Vendor with the Lowest Item Rate:** This table shows the vendor(s) offering the lowest price ('Item Rate') for the selected item(s) and filters. The 'Item Rate' is highlighted in light green to make it easy to spot the lowest price. This helps identify the most cost-effective source for the item(s) under consideration.
*   **Top 10 Vendors by Item Rate (Lowest to Highest):** This table lists the top 10 vendors based on their 'Item Rate' for the selected item(s) and filters, sorted from the lowest rate upwards. The 'Item Rate' column uses a green color gradient to visually represent the relative prices among these top vendors – darker green indicates a lower rate. Below the table, a bar chart visually compares the 'Item Rate' of these top 10 vendors, providing a quick overview of price variations.

### When a Specific Vendor is Selected:

*   **Items provided by [Selected Vendor]:** When you select a specific vendor, the dashboard lists the unique items that vendor has provided based on the other applied filters (Hospital and Parent Category), along with the 'Total Value' for each item. This helps understand the range of items a specific vendor supplies and their total contribution.

## Styling and Formatting

Numerical values in the results tables, especially the 'Item Rate', are styled (e.g., highlighted in green) to draw attention to key data points and make comparisons easier. The top 10 vendors table uses a color gradient on the 'Item Rate' to visually represent the price range.
```
